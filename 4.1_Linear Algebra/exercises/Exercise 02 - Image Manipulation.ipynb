```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image_array, show_info=True):
    cmap = "viridis"
    if len(image_array.shape)==2:
        cmap = "gray"
    
    if show_info:
        print(f"Image shape: {image_array.shape}")

    plt.imshow(image_array, cmap)
    plt.axis('off') 
    plt.show()

```

```python
image_path = "./assets/cat.jfif"
image = Image.open(image_path)

color_image = np.array(image)

show_image(color_image, True)
```

```python
grayscale_image = 0.2989 * color_image[:, :, 0] + 0.5870 * color_image[:, :, 1] + 0.1140 * color_image[:, :, 2]
grayscale_image = np.round(grayscale_image).astype(np.uint8)

show_image(grayscale_image)
```

```python
flipped_image_horizontally = np.fliplr(grayscale_image)
show_image(flipped_image_horizontally)
```

```python
def crop_image(image_array, start_row, end_row, start_col, end_col):
    cropped_image = image_array[start_row:end_row, start_col:end_col]
    show_image(cropped_image)
    return cropped_image
```

```python
cropped_image = crop_image(flipped_image_horizontally, 100, 400, 200, 500)
```

```python
final_image = Image.fromarray(cropped_image.astype(np.uint8))
final_image.save("./assets/cat_cropped.jpg")
```
